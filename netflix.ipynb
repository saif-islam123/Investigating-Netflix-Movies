```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
movies_1990s = netflix_df[(netflix_df['release_year'] >= 1990) & (netflix_df['release_year'] <= 1999) ]
most_frequent_duration = movies_1990s['duration'].mode().iloc[0]
duration = int(most_frequent_duration) 
short_movie_count = 0
movie_genre = movies_1990s[movies_1990s['genre'] == 'Action']
for label, row in movie_genre.iterrows():
    if row['duration'] <= 90:
        short_movie_count += 1
print(short_movie_count)
movie_duration = movies_1990s['duration']
plt.hist(movie_duration, bins = 20, edgecolor = 'black')
plt.xlabel('Duration')
plt.ylabel('Number of Movies ')
plt.title("Duration of movies in 1990s ")
plt.show()

```
